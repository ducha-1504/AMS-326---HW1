```python
#Create math function f(x)
def f1(x):
  return math.exp((-x**4)) - x**3  - math.cos(1-x**2)
```

Calculating Flop:
- math.exp((-x^4)): 7 + 1 + 3 = 11 FLOPs
- -x^3: 1 + 2 = 3 FLOPs
- -math.cos(1-x^2): 1+ 7 + 1 + 1 = 10 FLOPs
--> In total: 24 FLOPs

Method 1: Bisection Method

```python
#Check if [-1,1] is an appropriate inital bracket
def check_negative():
  a,b = -1,1
  results = 0
  while f1(-1) * f1(1) >= 0:
    raise ValueError("f(a) and f(b) must have opposite sign")
  return a, b
```

```python
#The Bisection Method
def BisectionMethod ():
  a, b = check_negative()
  TOL = 1e-4
  while (b - a)/2 > TOL:
    c = (b + a)/2
    if f1(c) == 0:
      return c
    elif f1(a) * f1(c) < 0:
      b = c
    else:
      a = c

  final_answer = (a + b) / 2
  return final_answer
```

Calculating FLOPs for Bisection Method:
- (b - a)/2: 1 + 7 = 8 FLOPs
- c = (b + a)/2: 1 + 7 = 8 FLOPs
- f1(c): 24 FLOPs
- f1(a) * f1(c): 24 + 1 + 24 = 49 FLOPs
--> In total: 8 + 8 + 24 + 49 = 89 FLOPs per iterations
- (a + b)/2: 1 + 7 = 8 FLOPs
- Assume k iterations, there are: 89k + 8 FLOPs

Method 2: Newton's Method

```python
#Finding derivative of function f(x)
def f2(x):
  return math.exp(-(x**4)) - 3 * x**2 - 2 * x * math.sin(1 - x**2)
```

```python
def Newton(x0, r):
  x1 = 0
  TOL = 1e-5
  while (x1 - r) > TOL or (r - x1) > TOL :
    x1 = x0 - (f1(x0)) / (f2(x0))
    x0 = x1
  return x1
```

Calculating FLOPs for Newton's Method:
- f2(x): 26 FLOPs
- f1(x0): 24 FLOPs
- f2(x0): 26 FLOPs
- while condition: 2 FLOPs
- x0 - (f1(x0)) / (f2(x0)): 1 + 24 + 26 + 7 = 58 FLOPs
- In total: 2 + 58 = 60 FLOPs
- Assume k iterations, there are 60k FLOPs



Method 3: Secant Method

```python
def Secant(x_secant0, x_secant1, r):
  TOL = 1e-4
  x = x_secant1
  while (x - r) > TOL or (r - x) > TOL:
    y0 = f1(x_secant0)
    y1 = f1(x_secant1)
    x = x_secant1 - (x_secant1 - x_secant0) * ((y1)/(y1 - y0))
    x_secant0 = x_secant1
    x_secant1 = x
  return x
```

Calculating FLOPs for the Secant Method:
- While condition: 2 FLOPs
- f1(x_secant0): 24 FLOPs
- f1(x_secant1): 24 FLOPs
- x = x_secant1 - (x_secant1 - x_secant0) * ((y1)/(y1 - y0)): 1 + 1 + 1 + 7 + 1 =  11 FLOPs
- In total: 2 + 24 + 24 + 11 = 61 FLOPs per iteration
- Assume k iterations, there are 61k FLOPs


Method 4: Monte Carlo Method

```python
def MonteCarlo(x_mon1, x_mon2):

  for iteration in range (5):
    best_x = None
    best_error = float('inf')

    for _ in range(10000):
      x = x_mon1 + (x_mon2 - x_mon1) * random.random()
      error = abs(f1(x))

      if error < best_error:
        best_error = error
        best_x = x

    #Shrink interval around best_x
    width = (x_mon2 - x_mon1) * 0.2
    x_mon1 = best_x - width/2
    x_mon2 = best_x + width/2

  return x
```

Calculating FLOPs for Monte Carlo Method:
- x = x_mon1 + (x_mon2 - x_mon1) * random.random(): 3 FLOPs
- error = abs(f1(x)): 24 + 1 = 25 FLOPs
- So, 28 FLOPs per inner loop
- 10000 samples iteration, so: 2.8 x 10^5 FLOPs
- Shrink interval: 11 FLOPs
- In total: 5 x (2.8 x 10^4 + 11) ~ 1.4 x 10^6 FLOPs
